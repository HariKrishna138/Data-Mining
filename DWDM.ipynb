```python
'''import pandas as pd'''
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("/content/my1.csv")
print(data.head())
sns.set(style = "white")
plt.figure(figsize=(6,4))
sns.histplot(data['Age'],kde = True,color='aqua')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.figure(figsize=(6,4))
sns.boxplot(x='Department',y='Salary',data=data,palette='Set2')
plt.title("Box Plot of Salary by Department")
plt.xlabel("Department")
plt.figure(figsize=(6,4))
avg_salary = data.groupby('Department')['Salary'].mean().reset_index()
sns.barplot(x='Department',y='Salary',data=avg_salary,palette='coolwarm')
plt.title("Bar Plot of Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Department',y='Salary',data=data,palette='Set2')
plt.title("Box Plot of Salary by Department")
plt.xlabel("Department")
plt.show()
```

```python
plt.figure(figsize=(6,4))
avg_salary = data.groupby('Department')['Salary'].mean().reset_index()
sns.barplot(x='Department',y='Salary',data=avg_salary,palette='coolwarm')
plt.title("Bar Plot of Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(6,4))
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=90,colors=['#ff9999','#66b3ff'])
plt.title("Gender Distribution")
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age',y='Salary',data=data,hue='Gender',palette='coolwarm')
plt.title("Scatter Plot of Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
import numpy as np
from scipy.spatial.distance import pdist,squareform
data = pd.DataFrame({
    'Height':[160,165,170,175],
    'Weight':[55,60,65,70],
},index=['A','B','C','D'])
distance_matrix = pd.DataFrame(squareform(pdist(data)),index=data.index,columns=data.index)
print("\nDissimilarity Matrix(Euclear Distance).")
print(distance_matrix)
```
